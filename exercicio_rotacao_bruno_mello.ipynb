```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rotate_point(x, y, theta=0):
    return x*np.cos(theta)-y*np.sin(theta), x*np.sin(theta)+y*np.cos(theta)
```

```python
def translate_point_x(x, y, trans=0):
    return x+trans, y
```

```python
def draw_figure(trans):
    center = (0+trans, 0)
    rot_point = rotate_point(0, -1, trans)
    point = rot_point[0]+center[0], rot_point[1]+center[1]
    circle1 = plt.Circle(center, 1, fill=False)
    plt.gca().add_patch(circle1)
    plt.plot(*point, 'ro')
    
draw_figure(0)
draw_figure(np.pi)
draw_figure(2*np.pi)
plt.axis('equal')
plt.show()
```
