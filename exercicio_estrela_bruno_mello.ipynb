```python
import matplotlib
import pylab as pl
import matplotlib.patches as patches
from matplotlib.path import Path
import numpy as np
from itertools import zip_longest
import matplotlib.pyplot as plt
```

```python
def merge(l1, l2):
    for i in zip_longest(l1, l2):
        for j in i:
            if j is not None:
                yield j
```

```python
def get_outter_points(r_o=1, rot_angle=0):
    return [(r_o*np.cos(2*np.pi*k/5+np.pi/2+rot_angle), 
             r_o*np.sin(2*np.pi*k/5+np.pi/2+rot_angle)) for k in range(5)]

def get_inner_points(r_i=0.5, rot_angle=0):
    return [(r_i*np.cos(2*np.pi*k/5+np.pi/2+2*np.pi/10+rot_angle), 
             r_i*np.sin(2*np.pi*k/5+np.pi/2+2*np.pi/10+rot_angle)) for k in range(5)]
```

```python
def plot_star(scale=1, rot_angle=0, transl_vect=(0,0), cis_x=0, inv_x=False):
    x_o, y_o = zip(*get_outter_points(scale, rot_angle))
    x_i, y_i = zip(*get_inner_points(scale/2, rot_angle))
    
    x = list(map(lambda i: i+transl_vect[0], merge(x_o, x_i)))
    y = list(map(lambda i: -i+transl_vect[1] if inv_x else i+transl_vect[1], merge(y_o, y_i)))
        
    x = list(map(lambda i: i[0]+cis_x*i[1], zip(x,y)))
    
    plt.plot(x+[x[0]], y+[y[0]])
```

```python
plot_star()
```

```python
plot_star(inv_x=True)
```

```python
plot_star(rot_angle=np.pi/7)
```

```python
plot_star(scale=4)
```

```python
plot_star(cis_x=1)
```

```python
plot_star(transl_vect=(1,2))
```
